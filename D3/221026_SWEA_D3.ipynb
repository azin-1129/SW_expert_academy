```python
# 12368. 24시간
# 4분 소요

T=int(input())

for t in range(1,T+1):
  time=0
  A,B=map(int,input().split())

  if A+B>=24:
    time=(A+B)-24
  else:
    time=A+B

  print(f'#{t} {time}')
```

```python
# 3431. 준환이의 운동관리
# 7분 소요

T=int(input())

for t in range(1,T+1):
  L,U,X=map(int,input().split())

  if X>U:
    print(f'#{t} {-1}') # 더하고 있음
  elif X<=L:
    print(f'#{t} {L-X}') # L-X 만큼 더 해야함
  else:
    print(f'#{t} {0}') # 적당함
```

```python
# 10505. 소득 불균형
# 5분 소요

T=int(input())

for t in range(1,T+1):
  N=int(input())

  moneys=list(map(int,input().split()))

  comp=sum(moneys)/N
  count=0

  for money in moneys:
    if money<=comp:
      count+=1

  print(f'#{t} {count}')
```

```python
# 4406. 모음이 보이지 않는 사람
# 9분 소요

T=int(input())
compare=['a','e','i','o','u']

for t in range(1,T+1):
  str_list=list(input())

  for comp in compare:
    for _ in range(str_list.count(comp)):
      str_list.remove(comp)

  print(f'#{t}',''.join(str_list))
```

```python
# 1217. 거듭 제곱
# 10분 소요

def recurse(X,Y):
  if Y==1:
    return X
  
  return X*recurse(X,Y-1)

for _ in range(10):
  t=int(input())
  x,y=map(int,input().split())

  print(f'#{t} {recurse(x,y)}')
```

```python
# 1215. 회문1
# 1시간 30분 소요. 회문 리스트에서 중복 배제하지 말걸 그랬음

for t in range(1,11):
  slice_range=int(input())

  graph=[]
  pali_list=[] # 회문 목록
  data=[]

  for _ in range(8):
    graph.append(list(input()))

  slice_half=slice_range//2 # 절반 길이

  if slice_range%2==0: # 짝수일 때
    for x in range(8): # 행 단위 회문 탐색
      #print(x,'번째 행 탐색중.')
      for i in range(0, 8-slice_range+1):
        data=graph[x][i:i+slice_range]
        #print(data,data[0:slice_half],'와',data[slice_half:slice_range],'비교중')
        if data[0:slice_half]==list(reversed(data[slice_half:slice_range])): #and data not in pali_list:
          #print(data,'는 회문')
          pali_list.append(data)
    
    for x in range(8): # 열 단위 회문 탐색
      #print(x,'번째 열 탐색중.')
      data_tmp=[]
      for j in range(8):
        data_tmp.append(graph[j][x]) # 각 열 삽입
        
      for i in range(0,8-slice_range+1):
        data=data_tmp[i:i+slice_range]
        #print(data,data[0:slice_half],'와',data[slice_half:slice_range],'비교중')
        if data[0:slice_half]==list(reversed(data[slice_half:slice_range])): #and data not in pali_list:
          #print(data,'는 회문')
          pali_list.append(data) # 열 데이터마다 회문 조사
  else: # 홀수일 때
    for x in range(8): # 행 단위 회문 탐색
      #print(x,'번째 행 탐색중.')
      for i in range(0, 8-slice_range+1):
        data=graph[x][i:i+slice_range]
        #print(data,data[0:slice_half],'와',data[slice_half+1:slice_range],'비교중')
        if data[0:slice_half]==list(reversed(data[slice_half+1:slice_range])): #and data not in pali_list:
          #print(data,'는 회문')
          pali_list.append(data)
    
    for x in range(8): # 열 단위 회문 탐색
      #print(x,'번째 열 탐색중.')
      data_tmp=[]
      for j in range(8):
        data_tmp.append(graph[j][x]) # 각 열 삽입
        
      for i in range(0,8-slice_range+1):
        data=data_tmp[i:i+slice_range]
        #print(data,data[0:slice_half],'와',data[slice_half+1:slice_range],'비교중')
        if data[0:slice_half]==list(reversed(data[slice_half+1:slice_range])): #and data not in pali_list:
          #print(data,'는 회문')
          pali_list.append(data) # 열 데이터마다 회문 조사

  print(f'#{t} {len(pali_list)}')
```

```python
# 13218. 조별과제
# 10분 소요. 최소 팀 인원 3명인거 봐놓고 2명까지 팀으로 짰네

T=int(input())

for t in range(1,T+1):
  N=int(input())

  if N<3:
    print(f'#{t} {0}')
  else:
    print(f'#{t} {N//3}')
```
